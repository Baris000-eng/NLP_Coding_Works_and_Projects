```python
# We can use the PyPDF2 library to 
# read text data from a PDF file.

# Important Note: Not all PDFs have text that 
# can be extracted.

# Some PDFs are created through scanning,
# instead of being exported from a text 
# editor like Word. These scanned PDFs 
# consist of image files, which makes much 
# harder to extract the text from them.
# Text extraction from the scanned PDFs 
# often requires a specialized software.

# The PyPDF2 library is made to extract text 
# from PDF files directly created from a word 
# processor, instead of from images.
```

```python
import PyPDF2
# 'rb' mode is utilized to read the pdf file in binary mode.
file = open('declaration.pdf', mode='rb')

pdf_reader = PyPDF2.PdfReader(file)
```

```python
len(pdf_reader.pages) # It will get the number of pages in the pdf file.
```

```python
# It will get the first page of the pdf file.
first_page = pdf_reader.pages[0] 

# It will extract the text from the first page of the PDF file,
# and print the text from that page.
print(first_page.extract_text()) 
```

```python
second_page = pdf_reader.pages[1] # It will get the second page of the PDF file.
print(second_page.extract_text()) # It will extract the text from the second page of the PDF file and prints it.
```

```python
third_page = pdf_reader.pages[2] # It will get the 3rd page of the PDF file.
extracted_text = third_page.extract_text() # It will extract the text from the 3rd page.
print(extracted_text) # It will print the extracted text to the console.
```

```python
fourth_page = pdf_reader.pages[3]
fourth_page_text = fourth_page.extract_text()
print(fourth_page_text)
```

```python
fifth_page = pdf_reader.pages[4]
fifth_page_text = fifth_page.extract_text()
print(fifth_page_text)
```

```python
# The below call will throw an index error because the PDF document includes 5 pages 
# and the 5th index belongs to the 6th page, which is non-existent in the acceptable 
# sequence of page indices. This sequence starts at 0 and ends at 4.
print(pdf_reader.pages[5].extract_text())
```

```python
# Important Note: We cannot directly write Python string text onto a PDF file due to the differences between a 
# single string type of Python and the diversity of placements, fonts, and other parameters that a PDF file might 
# have. Instead, we can copy PDF pages and append them to the end.
```

```python
my_file = open('declaration.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(my_file)

initial_page = pdf_reader.pages[0]
pdf_writer = PyPDF2.PdfWriter()
pdf_writer.add_page(initial_page)
```

```python
pdf_output = open('resume.pdf', 'wb')
```

```python
pdf_writer.write(pdf_output)
```

```python
pdf_output.close()
my_file.close()
```

```python
# grabbing all the text from a PDF file
declaration_file = open('declaration.pdf', 'rb')

pdf_text = list()
text_of_pdf = str()
pdf_reader_object = PyPDF2.PdfReader(declaration_file)
for page_index in range(0, len(pdf_reader_object.pages)):
    page = pdf_reader_object.pages[page_index]
    extracted_text = page.extract_text()
    text_of_pdf += extracted_text
    pdf_text.append(extracted_text)
    
declaration_file.close()
```
